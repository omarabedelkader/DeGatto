# Sentence Level

```python
df_log_reg = pd.read_excel("/content/drive/MyDrive/DeGatto Project/EDA + Models/Sentiment Output.xlsx", sheet_name = "Logistic Regression")
df_mnb = pd.read_excel("/content/drive/MyDrive/DeGatto Project/EDA + Models/Sentiment Output.xlsx", sheet_name = "Multinomial Naive Bayes")
df_svm = pd.read_excel("/content/drive/MyDrive/DeGatto Project/EDA + Models/Sentiment Output.xlsx", sheet_name = "SVM")
```

```python
df = df_log_reg[(df_log_reg['Test Size'] == 0.1) & (df_log_reg['Vectorization Method'] == 'Count Vectorizer')]

y1 = df[df['C value'] == 0.5].set_index('Max Iterations')
y2 = df[df['C value'] == 1.0].set_index('Max Iterations')
y3 = df[df['C value'] == 1.5].set_index('Max Iterations')
y4 = df[df['C value'] == 2.0].set_index('Max Iterations')

plotdata_1 = pd.DataFrame({
  "C Value 0.5":round(y1['Accuracy Test'], 2),
  "C Value 1.0":round(y2['Accuracy Test'], 2),
  "C Value 1.5":round(y3['Accuracy Test'], 2),
  "C Value 2.0":round(y4['Accuracy Test'], 2)
  }
)

plotdata_1.plot(figsize = (8, 5))

plt.title('LR Count Vectorizer: Sentence-Level Accuracy Test with Test size 10%')
plt.xlabel("Max Iterations")
plt.ylabel("Accuracy Test")
  #plt.xticks()
  #plt.yticks(np.arange(0.83, 0.95, 0.1))
```

```python
df = df_mnb[df_mnb['Vectorization Method'] == 'TF-IDF']

y0 = df[df['Test Size'] == 0.10].set_index('Alpha')
y1 = df[df['Test Size'] == 0.15].set_index('Alpha')
y2 = df[df['Test Size'] == 0.20].set_index('Alpha')
y3 = df[df['Test Size'] == 0.25].set_index('Alpha')
y4 = df[df['Test Size'] == 0.30].set_index('Alpha')

plotdata_1 = pd.DataFrame({
      "Test Size 0.10":y0['Accuracy Test'],
      "Test Size 0.15":y1['Accuracy Test'],
      "Test Size 0.20":y2['Accuracy Test'],
      "Test Size 0.25":y3['Accuracy Test'],
      "Test Size 0.30":y4['Accuracy Test'],
      }
)

plotdata_1.plot(figsize = (8, 5))

plt.title('Multinomial Naive Bayes TF-IDF: Sentence-Level Accuracy Test')
plt.xlabel("Alpha")
plt.ylabel("Accuracy Test")
plt.ylim(0.75, 0.85)
plt.show()
```

```python
df = df_svm[df_svm['Vectorization Method'] == 'Count Vectorizer']

y0 = df[df['Test Size'] == 0.10].set_index('C Value')
y1 = df[df['Test Size'] == 0.15].set_index('C Value')
y2 = df[df['Test Size'] == 0.20].set_index('C Value')
y3 = df[df['Test Size'] == 0.25].set_index('C Value')
y4 = df[df['Test Size'] == 0.30].set_index('C Value')

plotdata_1 = pd.DataFrame({
      "Test Size 0.10":y0['Accuracy Test'],
      "Test Size 0.15":y1['Accuracy Test'],
      "Test Size 0.20":y2['Accuracy Test'],
      "Test Size 0.25":y3['Accuracy Test'],
      "Test Size 0.30":y4['Accuracy Test'],
      }
)

plotdata_1.plot(figsize = (8, 5))

plt.title('Linear SVC Count Vectorizer: Sentence-Level Accuracy Test')
plt.xlabel("C Value")
plt.ylabel("Accuracy Test")
plt.ylim(0.91, 0.95)

plt.show()
```

```python
'''
Bar plot showing the best accuracies for all the models in Sentence Level
'''

data = { 'Multinomial Naive Bayes':0.85, 'Logistic Regression': 0.94, 
        'LinearSVC':0.94, 'LSTM':0.95,
        }
models = list(data.keys())
accuracies = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.barh(models, accuracies, color =['#658CBB', '#D3494E', '#B879AC', '#8eb54e'])
 
plt.xlabel("Models Tested")
plt.ylabel("Accuracies")
plt.title("Accuracies of the four models tested in Sentence-Level")
plt.show()
```
