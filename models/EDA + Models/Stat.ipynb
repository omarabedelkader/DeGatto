```python
from scipy import stats
import pandas as pd# import the data
df= pd.read_excel("Sentiment Output.xlsx")
CountVectorizer = df[(df['Vectorization Method'] == 'Count Vectorizer')]
TFIDF = df[(df['Vectorization Method'] == 'TF-IDF')]

# homogeneity
stats.levene(CountVectorizer['Accuracy Test'], TFIDF['Accuracy Test'])

# Shapiro-Wilk test for normality
stats.shapiro(CountVectorizer['Accuracy Test'])
stats.shapiro(TFIDF['Accuracy Test'])
```

```python
stats.ttest_ind(CountVectorizer['Accuracy Test'], TFIDF['Accuracy Test'])
```
